# Ecosystems Display Example

This notebook demonstrates how the `Ecosystems` class renders as an HTML table in Jupyter notebooks.

```python
import ee
from google.auth import default

# Authenticate and initialize Earth Engine
credentials, _ = default(scopes=[
    'https://www.googleapis.com/auth/earthengine',
    'https://www.googleapis.com/auth/cloud-platform'
])
ee.Initialize(credentials=credentials, project='goog-rle-assessments')
```

```python
from rle_python_gee import Ecosystems
```

## Load a Raster Ecosystem Asset

When you display an `Ecosystems` object in a Jupyter cell, it automatically renders as an HTML table showing the asset metadata.

```python
# Load a raster ecosystem asset using asset_id string
raster_ecosystems = Ecosystems(
    'projects/goog-rle-assessments/assets/myanmar/mm_ecosys_v7b',
)
raster_ecosystems
```

## Load a Vector Ecosystem Asset

Vector assets (FeatureCollections) display feature count and property names.

```python
# Load a vector ecosystem asset using asset_id string with GET Level 3 and Level 4 columns
vector_ecosystems = Ecosystems(
    'projects/goog-rle-assessments/assets/columbia/GETCol',
    get_level3_column='EFG1',
    get_level456_column='Glob_Typol'
)
vector_ecosystems
```

## Create Ecosystems from an ee.FeatureCollection

You can also pass an `ee.FeatureCollection` or `ee.Image` directly to the `Ecosystems` constructor.

```python
# Create an ee.FeatureCollection and pass it directly to Ecosystems
fc = ee.FeatureCollection('projects/goog-rle-assessments/assets/columbia/GETCol')
ecosystems_from_fc = Ecosystems(
    fc,
    get_level3_column='EFG1',
    get_level456_column='Glob_Typol'
)
ecosystems_from_fc
```

## Access Records as a DataFrame

Use the `dataframe` property to get all records as a pandas DataFrame.

```python
# Get all records as a pandas DataFrame
df = vector_ecosystems.functional_group_dataframe()
df
```
