Example of using GeoJSON data sources

# Ecosystems Map

```python
from rle_python_gee.ecosystems import Ecosystems
from rle_python_gee.aoo import make_aoo
```

```python
ecosystems = Ecosystems.from_geojson('../tests/test_data/null_island.geojson')
ecosystems.load()
ecosystems
```

```python
ecosystems.to_map()
```

# AOO

```python
aoo = make_aoo(ecosystems).compute()
aoo
```

```python
aoo.to_map()
```

```python
from lonboard import Map

Map(layers=ecosystems.to_layer() + aoo.to_layer())
```
